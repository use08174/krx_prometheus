````python
# GPT 호출 함수 (질답 생성 시)
def to_GPT(prompt):
    response = client.chat.completions.create(
        model="gpt-4o-mini",
        messages=[
            {"role": "system", "content": "You are an assistant generating high-quality Q&A pairs for fine-tuning in finance."},
            {"role": "user", "content": prompt}
        ],
        max_tokens=3000,
        temperature=0.7
    ).to_dict()
    return response

# 형식 검증 함수
def validate_response_format(response, expected_structure):
    try:
        dict_test = eval(response)

        # Question 뒤에 문자열이 있는지 확인
        question_present = re.search(r'Question:\s*(.+?)\s*\|', response)
        
        # Option 뒤에 문자열이 있는지 확인
        options_present = re.search(r'Option:\s*\{(.+?)\}\s*\|', response)
        
        # Answer 뒤에 문자열이 있는지 확인
        answer_present = re.search(r'Answer:\s*(.+?)\s*\|', response)
        
        # Explanation 뒤에 문자열이 있는지 확인
        explanation_present = re.search(r'Explanation:\s*(.+)', response)
    
        return bool(question_present and options_present and answer_present and explanation_present)
    except:
        return False

# 응답 파싱 함수
def parse_response(response):
    """응답 텍스트에서 Question, Option, Answer를 추출하여 딕셔너리로 반환."""
    # Question 추출
    question_match = re.search(r'Question:\s*(.+?)\s*\|', response)
    question = question_match.group(1) if question_match else None

    # Option 추출
    option_match = re.search(r'Option:\s*\{(.+?)\}\s*\|', response)
    option = option_match.group(1) if option_match else None

    # Answer 추출
    answer_match = re.search(r'Answer:\s*(.+)', response)
    answer = answer_match.group(1) if answer_match else None

    # Answer 추출
    explanation_match = re.search(r'Explanation:\s*(.+)', response)
    explanation = explanation_match.group(1) if explanation_match else None

    return {
        'Question': question,
        'Option': option,
        'Answer': answer,
        'Explanation': explanation
    }

# 기업 정보와 설명을 무작위로 선택
def get_domestic_corp_external_info():
    corp_code = random.choice(list(info_dict.keys()))
    target_corp_desc = random.choice(info_dict[corp_code])
    corp_name = name_and_ticker_df.loc[name_and_ticker_df['corp_ticker'] == corp_code, 'corp_name'].values[0]
    
    return {
        'corp_code': corp_name,
        'corp_desc': target_corp_desc
    }

# 주가 예측용 외부 정보 가져오기
def get_stock_estimation_external_info():
    return get_price_external_info()

# Q&A 세트 생성 함수
def gen_QA(type):
    gen_QA_prompt = '''
    You must create high quality QA datasets for benchmarking financial LLMs.
    All the generation must be written in Korean.
    You must generate detailed explanation for each solution.
    ** You must strictly follow the given structure while generating response. **
    '''
    
    # 기본 구성 요소 설정
    sample_reference = f"Sample Reference: {sample_dict[type]}\n"
    response_structure = f"Response Structure: {structure_dict[type]}\nYou must follow the given structure."
    prompt_content = f"Prompt: {prompt_dict[type]}"
    
    # 외부 정보 설정
    external_info = None
    if type == '국내기업':
        external_info = get_domestic_corp_external_info()
    elif type == '주가예측':
        external_info = get_stock_estimation_external_info()
        save_price_table.append(external_info)
    elif type == '재무회계':
        selected_task = random.choice(accounting_topics)
        gen_QA_prompt += f"\n\n**Generate a question related to: {selected_task}.**"
    elif type == '금융시장':
        selected_task = random.choice(fin_market_topics)
        gen_QA_prompt += f"\n\n**Generate a question related to: {selected_task}.**"
        gen_QA_prompt += "\n\n**Ensure that the correct answer is limited to only one option.**"
    elif type == '금융에이전트':
        selected_task = random.choice(fin_agent_topics)
        gen_QA_prompt += f"\n\n**Generate a question related to: {selected_task}.**"

    # 최종 프롬프트 생성
    gen_QA_prompt += sample_reference + response_structure + prompt_content
    if external_info:
        gen_QA_prompt += f"\nExternal Information: {external_info}"

    # 응답 생성 및 검증
    for attempt in range(3):  # 최대 3회 시도
        response = to_GPT(gen_QA_prompt)
        response_content = response['choices'][0]['message']['content']

        # validate_response_format 함수에 type 전달
        if validate_response_format(response_content, structure_dict[type]):
            return response_content
        else:
            print('출력 구조가 structure dict와 다릅니다. 다시 시도합니다...')
            # 형식이 맞지 않는 경우 재시도 요청
            gen_QA_prompt += "\n\n**The previous response did not match the required structure. Please regenerate strictly following the given structure.**"

    # 시도 후에도 형식이 맞지 않는 경우 마지막 응답 반환
    return response_content
````
